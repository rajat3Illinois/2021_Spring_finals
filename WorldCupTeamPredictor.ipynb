```python
players_df = pd.read_csv("players_20.csv", index_col="sofifa_id")
players_df
```

```python
team_details = team_details.set_index("Nation")
team_details
```

```python
grouped_players_df = players_df.groupby('nationality')['overall'].agg(['median', 'count'])
grouped_players_df.head(30)
```

```python
team_player_details = pd.merge(team_details, grouped_players_df, left_on='Nation', right_index=True, how='left').fillna(0)
team_player_details
```

```python
team_player_details['Rank'] = team_player_details['Rank']
team_player_details
```

```python
max_rank = team_player_details.max()['Rank']
max_rank
```

```python
min_rank = team_player_details.min()['Rank']
min_rank
```

```python
team_player_details['Normalize_Rank'] = (max_rank - team_player_details['Rank']) / (max_rank - min_rank)
team_player_details
```

```python
team_player_details['Above_Median_Count'] = team_player_details['count']/ 2 
team_player_details['Above_Median_Count'] =team_player_details['Above_Median_Count'].astype("int32")
team_player_details
```

```python
fetched_team = team_player_details[team_player_details['Normalize_Rank'].gt(0.4)]
fetched_team
```

```python
teams_qualified = team_player_details.loc[(team_player_details['Normalize_Rank'].gt(0.4)) & 
((team_player_details['Squad_Size'] * 1.5) < team_player_details['Above_Median_Count'])]
teams_qualified
```

```python
team_player_details['Total Value'] = team_player_details['Total Value'].replace({'\$':''}, regex = True)
team_player_details['Total Value'] = team_player_details['Total Value'].replace({'m':''}, regex = True)
team_player_details['Total Value'] = team_player_details['Total Value'].replace({'bn':''}, regex = True)
team_player_details['Total Value'] = team_player_details['Total Value'].replace({'Th': ''}, regex=True)
team_player_details
```

```python
team_player_details['Total Value']=team_player_details['Total Value'].astype(float)
```

```python
different_teams_qualified = team_player_details.loc[(team_player_details['Normalize_Rank'].gt(0.4)) & 
(team_player_details['Total Value'].gt(150))].reset_index()
different_teams_qualified
```

```python
teams_qualified = team_player_details.loc[(team_player_details['Normalize_Rank'].gt(0.3)) & 
((team_player_details['Squad_Size'] * 0.5) < team_player_details['Above_Median_Count'])]
teams_qualified
```

```python
different_values = teams_qualified[~teams_qualified.isin(different_teams_qualified)].dropna()
different_values
```

```python
UEFA_Higher_Pot = team_player_details[team_player_details['Pots'].isin([1, 2]) 
                                      & team_player_details['Normalize_Rank'].gt(0.4) & 
((team_player_details['Squad_Size'] * 1.5) < team_player_details['Above_Median_Count']) &
                                     (team_player_details['Total Value'].gt(150))]
UEFA_Higher_Pot
```

```python
UEFA_Lower_Pot = team_player_details[~team_player_details['Pots'].isin([1, 2]) 
                                      & team_player_details['Normalize_Rank'].gt(0.4) & 
((team_player_details['Squad_Size'] * 1.5) < team_player_details['Above_Median_Count']) &
                                     (team_player_details['Total Value'].gt(100))]
UEFA_Lower_Pot
```
